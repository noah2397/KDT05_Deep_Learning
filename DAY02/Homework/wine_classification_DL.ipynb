```python
import pandas as pd
import torch
import torch.nn as nn              
import torch.nn.functional as F       
import torch.optim as optim      
# 134~140p의 코드를 그대로 따라침 

      
df=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv", sep=";")
data=torch.tensor(df[df.columns[:-1]].values).float()
target=torch.tensor(df[df.columns[-1]].values).long()
data.dtype, target.dtype
```

```python
target = target-3
```

```python
model = nn.Sequential(
    nn.Linear(11, 7)
)
```

```python
optimizer=optim.SGD(model.parameters(), lr=1e-4)
for epoch in range(10000):
    hypothesis = model(data)
    print(target.shape)
    cost = F.cross_entropy(hypothesis, target)
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
    print(f'Epoch {epoch:4d} Cost: {cost.item():.6f}')
```

```python
count=0
for i in range(data.shape[0]):
    if target[i] == F.softmax(model(data[i])).max(dim=0)[1] :
        count+=1
print(count/data.shape[0])
```
